#### Exercise 01. Basic data analysis with native Python structures.  
Create a list with the grades of 5 students: [8.5, 9.2, 7.8, 8.9, 10].

```python
# TODO
calificaciones = [8.5, 9.2, 7.8, 8.9, 10]

promedio = sum(calificaciones) / len(calificaciones)

print("El promedio de las calificaciones es:", promedio)
```

#### Exercise 01. What is the most expensive house in the entire dataset? (★☆☆)

Print the address and price of the selected house. For example:

`The house located at Calle del Prado, Nº20 is the most expensive, and its price is 5000000 USD.`

```python
# TODO
import pandas as pd

df = pd.read_csv('assets/real_estate.csv', sep=';')

most_expensive = df.loc[df["price"].idxmax()]

print(
    f"La casa con dirección en {most_expensive['address']} "
    f"es la más cara y su precio es {most_expensive['price']} USD"
)
```

#### Exercise 07. What is the average price in the population of "Arroyomolinos (Madrid)"? (★★☆)

#### Exercise 08. Plot the histogram of prices for the population of "Arroyomolinos (Madrid)" and explain what you observe (★★☆)

Print the histogram of the prices and write a brief analysis of the plot in the Markdown cell.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el dataset
df = pd.read_csv('assets/real_estate.csv', sep=';')

# Filtrar por población
arroyo_prices = df[df["level5"] == "Arroyomolinos (Madrid)"]["price"]

# Trazar el histograma
plt.hist(arroyo_prices, bins=20)
plt.title("Histograma de precios en Arroyomolinos (Madrid)")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()

```

El histograma muestra que la mayoría de las viviendas en Arroyomolinos (Madrid)
se concentran en un rango de precios medio, con una frecuencia mayor en los
valores intermedios.

Se observa una ligera asimetría hacia la derecha, lo que indica la presencia
de algunas viviendas con precios más altos que elevan el promedio, aunque son
menos frecuentes.

Esto sugiere que el mercado inmobiliario de la población está dominado por
viviendas de precio medio, con algunos inmuebles de alto valor que actúan como
valores extremos.
